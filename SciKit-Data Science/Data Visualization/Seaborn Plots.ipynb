```python
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import datasets

import pandas as pd
```

```python
diabetes_datasets = datasets.load_diabetes()
diabetes_datasets
```

```python
columns = diabetes_datasets["feature_names"]
diabetes_df = pd.concat([pd.DataFrame(diabetes_datasets["data"], columns=columns), pd.DataFrame(diabetes_datasets["target"], columns=["target"])], axis=1)
diabetes_df
```

```python
sns.lineplot(data=diabetes_df, x="age", y="target")
```

```python
plt.title("Age Distribution")
age = sns.displot(diabetes_df.age)
```
