```python
def sliding_window(sentence, context_length):
    count = {}
    sentence = word_tokenize(sentence)
    for i, word in enumerate(sentence):
        count[word] = sentence[i - context_length: i+1] + sentence[i+1: i + context_length+1]
        # get the focus word positioning

    return count

sliding_window(documents.review.iloc[0], 3)
# word_tokenize(documents.review.iloc[0])
```
