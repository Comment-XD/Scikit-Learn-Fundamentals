```python
def box_plot(smoker_data, non_smoker_data, colors=["lightgreen", "pink"]):
    fig, ax = plt.subplots(int(len(smoker_data.columns) / 2), 2, figsize=(10, 40))
    
    for i, col in enumerate(smoker_data.columns):
        if col == "smoking":
            break
        j = i % 2
        ax[i % 11, j].set_title(col)
        ax[i % 11, j].yaxis.grid(True)
        data = [smoker_data[col], non_smoker_data[col]]
        boxplot = ax[i % 11, j].boxplot(data, labels=[f"smoker {col}", f"nonsmoker {col}"], patch_artist=True, notch=True)
            
        for patch, color in zip(boxplot["boxes"], colors):
            patch.set_facecolor(color)
    plt.show()
```

```python
box_plot(smokers_data, non_smokers_data)
```
